```python
import os
import requests
from bs4 import BeautifulSoup

def webscrap(url_list, output_directory):

    # Verifica se o diretório de saída existe, senão cria
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
        
    for url in url_list:
        # Fazendo a requisição GET
        response = requests.get(url)
        
        # Verificando se a requisição foi bem-sucedida
        if response.status_code == 200:
            # Criando o objeto BeautifulSoup
            soup = BeautifulSoup(response.content, "html.parser")
    
            # criando o nome do arquivo .txt de saída
            text_file = url.split("https://bg3.wiki/wiki/")[-1].lower()
            
            # Criando um arquivo para escrever
            with open(os.path.join(output_directory, f"{text_file}.txt"), "w", encoding="utf-8") as file:
                # Encontrando todos os cabeçalhos h2, h3, h4, etc.
                headers = soup.find_all(lambda tag: tag.name.startswith('h') and len(tag.name) == 2)
                
                # Iterando sobre os cabeçalhos
                for header in headers:
                    # Escrevendo o texto do cabeçalho no arquivo
                    file.write(header.get_text(strip=True) + "\n")
                    
                    # Encontrando os elementos irmãos até o próximo cabeçalho e escrevendo no arquivo
                    next_sibling = header.find_next_sibling()
                    while next_sibling:
                        if next_sibling.name and next_sibling.name.startswith("h"):
                            break
                        if next_sibling.name in ["p", "ul", "ol"]:
                            file.write(next_sibling.get_text(strip=True) + "\n")
                        next_sibling = next_sibling.next_sibling
                    
                    # Adiciona uma linha em branco entre os cabeçalhos
                    file.write("\n")
            
            print (f"Conteúdo salvo no arquivo '{text_file}.txt'")
            output = f"Todo conteúdo foi salvo!"
        else:
            output = f"Erro ao acessar a página: {url}"

    return output
```
